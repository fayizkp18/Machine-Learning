# SEABORN

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Seaborn uses fewer syntax and has stunning default themes

```python
#https://seaborn.pydata.org/
```

```python
#!PIP INSTALL SEABORN
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

# Hue,  Style and Size

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time', size= 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'size')
```

```python
tips['size'].value_counts()
```

# Line plot

```python
#from seaborn import relplot
sns.set(font_scale=1)
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = tips, height = 5, aspect =4)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line',  data = tips, label = 'Tip', height = 6, aspect = 2)
plt.legend(loc='right')
```

```python
sns.set(font_scale=0.75)
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line',hue = 'smoker',  data = tips,height = 6, aspect = 2)
plt.legend(loc='right')
```

# Subplot

```python
tips.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'smoker', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', data = tips, row = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', data = tips, col_wrap=3)
```

# Categorical Data Ploting 

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, jitter = False)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'size')
```

```python
sns.catplot(x = 'smoker', y = 'tip', data = tips, order = ['No', 'Yes'])
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex')
```

# Bar plot

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'bar', data = tips, hue = 'sex')
```

# Point Plot

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'point', data = tips, hue = 'smoker')
```

# Joint Plot 

```python
x = tips['total_bill']
y = tips['tip']
```

```python
sns.jointplot(x, y)
```

```python
sns.jointplot(x, y, kind = 'hex')
```

# Pair plot

```python
sns.pairplot(tips)
```
