```python
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Read comma separated data
data = np.loadtxt(os.path.join('Data', r'C:\Users\USER\Downloads\ex1data1.txt'), delimiter=',')
X, Y = data[:, 0], data[:, 1]
m = X.size  # number of training examples
print(X.shape)
```

```python
print('Plotting Data ...\n')
def plotData(x, y):
    plt.plot(x, y, 'ro', ms=10, mec='k')
    plt.ylabel('Profit in $10,000')
    plt.xlabel('Population of City in 10,000s')
    fig = plt.figure()
```

```python
plotData(X ,Y)
```

```python
print("The first five samples {}".format(X[:5]))
print("The first five targets {}".format(Y[:5]))
X= X.reshape(-1, 1)
Y= Y.reshape(-1, 1)
print("The number of samples in train set is {}".format(X.shape))
print("The number of samples in test set is {}".format(Y.shape))
lr = LinearRegression()
lr.fit(X, Y)
```

```python
y_pred = lr.predict(X)
plt.scatter(X, Y, color ='r')
plt.plot(X, y_pred, color ='b')
```
